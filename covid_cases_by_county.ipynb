```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('county-cumulative.csv', delimiter=',')
df.head()
```

```python
df.plot(x="County", y=["Cumulative Cases", "Population"])
```

```python
cc_column = df["Cumulative Cases"]
type(cc_column)

cc_column.plot(kind="hist")
```

```python
top_5 = df.sort_values(by="Cumulative Cases", ascending=False).head()

top_5.plot(x="County", y="Cumulative Cases", kind="bar", rot=5)
```

```python
top_cases = df[df["Cumulative Cases"] > 5000].sort_values("Cumulative Cases", ascending=False)

top_cases.plot(x="County", y=["Cumulative Cases", "Population"], kind="bar", rot=5)
```
